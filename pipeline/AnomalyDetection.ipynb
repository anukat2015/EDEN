```python
plt.figure(figsize=(20,10))
cmap = plt.get_cmap('jet')
colors = cmap(np.linspace(0, 1.0, len(s[:10])))
# for (name, group), color in zip(data_merge.groupby('event'), colors):
ts = {}
for (name, group), color in zip(s[:10], colors):
    group['first-published_dt'] = group['first-published'].apply(pd.to_datetime)
    group.index = group['first-published_dt']
    event_data = group.resample('24h',how={'first-published_dt':'count'})
    if str(name) in ['533.0', '472.0', '312.0', '519.0']:
        color = 'lightgrey'
        plt.plot(event_data, label=str('Cluster: ') + str(name), c=color, alpha=0.5)
    else:
        plt.plot(event_data, label=str('Cluster: ') + str(name), c=color)
    ts[name] = event_data['first-published_dt'].max()
    
plt.legend()
plt.ylabel('Article Count')
plt.xlabel('Article (published-time)')
```
